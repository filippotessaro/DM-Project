```python
import random
```

```python
def generateBasket():
    which = random.randint(0, 1)
    if which == 0:

        # 'official' random products

        n = len(products)
        quanti = n * (n + 1) * 0.5

        prdist = [(n - i)/quanti for i in range(len(products))]    
        extract = list(range(1, n + 1))    
        official = np.random.choice(extract, p=prdist)

        prod = random.sample(products, official)

    else:
        # AR Based

        probs = [l[2] for l in lista]
        tot = sum(probs)
        probs = [p/tot for p in probs]
        scelta = range(len(lista))
        scelta = [*scelta]

        index = np.random.choice(scelta, p=probs)

        AR = lista[index]

        n = len(AR[0])
        quanti = n * (n + 1) * 0.5

        extract = list(range(1, n + 1))  
        official = np.random.choice(extract)

        prod = random.sample(AR[0], official)

        yprob = probs[index]
        included = np.random.binomial(size=1, n=1, p=yprob)[0]

        if included == 1:
            prod.append(AR[1][0])

        prod = list(set(prod))

    return prod
```

```python
sintetico = []
for _ in range(10000):
    sintetico.append(generateBasket())
```
