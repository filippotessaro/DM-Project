```python
f = open("./data/groceries.csv")
cnt = f.read()
```

```python
csv = open("./data/groceries_aligned.csv", "w")
```

```python
rows = cnt.split("\n")

mmax = 0
for r in rows:
    mmax = max([len(r.split(",")), mmax])
    
for i in range(mmax - 1):
    csv.write("Product" + str(i+1) + ",")

csv.write("Product" + str(mmax))
csv.write("\n")

for r in rows:
    tokens = r.split(",")
    diff = mmax - len(tokens)
    
    toadd = ""
    for _ in range(diff):
        toadd += ","
    
    final = r + toadd + "\n"
    csv.write(final)
    
csv.close()
```
